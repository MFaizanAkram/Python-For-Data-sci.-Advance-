# 16:- Splitting a string into multiple columns:

```python
# Importing libraries
import pandas as pd

df = pd.DataFrame({'name':['Muhammad Akram','Muhammad Zeeshan','Ali Hamza','Faizan Akram'],
                   'location':['Sialkot, Pakistan', 'Brussels, Belgium','Faisalabad, Pakistan','Gujranwala, Pakistan']})
df
```

```python
# Splitting a column into two columns
df.name.str.split(' ',expand=True) 
```

```python
# Adding those splits into new columns
df[['first_name','last_name']] = df.name.str.split(' ',expand=True)
df 
```

- Expanding Location:

```python
df.location.str.split(' ',expand=True)
```

```python
df[['City','Country']] = df.location.str.split(' ',expand=True)
df
```

```python
# Refine data manipulation
df = df[['first_name','last_name','City','Country']]
df
```

# 23:- Reshaping a Data-frame:

```python
fasla = pd.DataFrame([['12345', 100,200,200], ['34567',400,500,600],['67890',700,800,900]],
                     columns=['zip','factory','warehouse','retail'])
fasla.head()
```

```python
import matplotlib.pyplot as mlt
import seaborn as sns
sns.barplot(x='zip', y='factory', data=fasla)
```

```python
fasla_long = fasla.melt(id_vars='zip', var_name='location_type', value_name='distance')
fasla_long
```

```python
sns.barplot(x='zip', y='distance', hue='location_type', data=fasla_long)
```
